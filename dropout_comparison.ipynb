#Dropout 0.1

```python
import matplotlib.pyplot as plt

# Provided test metrics
accuracy = 0.9771309494972229
precision = 0.9791666865348816
recall = 0.9771309494972229
auc = 0.9983406066894531
f1_score = 0.9781477588149342

# Metrics and corresponding values
metrics = ['Accuracy', 'Precision', 'Recall', 'AUC', 'F1 Score']
values = [accuracy, precision, recall, auc, f1_score]

# Plotting the bar diagram
plt.bar(metrics, values, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('CNN architecture using Dropout 0.1')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.ylim(0.8, 1.0)  # Set y-axis limits for better visualization

# Display the values on top of the bars
for i, value in enumerate(values):
    plt.text(i, value + 0.002, f'{value:.4f}', ha='center')

# Show the plot
plt.show()
```
